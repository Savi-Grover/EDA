```python
# read country code foreign table
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# combine main data and foreign table data : df_country and df

final_df=pd.merge(df,df_country, on='Country Code', how='left')  
# on is paarameter to merge , how= join type
final_df
# last col country added
```

```python
final_df.head(2)
```

```python
# inplace of info , we can use dtype to find datatype of features
final_df.dtypes
```

```python
# observe columns
final_df.columns
```

```python
# analayse how many records related to each country in dataset
final_df.Country.value_counts()
```

```python
# based on above, zomato is having main market in India
# second most in US
```

```python
# storing the sorted order by index
country_names=final_df.Country.value_counts().index
country_names
```

```python
# storing sorted values of records by counts
country_values=final_df.Country.value_counts().values
country_values
```

```python
## plot a pie chart using above 2
plt.pie(country_values,labels=country_names)
```

```python
# very unclear plot
# so we fiter top 5 countries where zomato is used
plt.pie(country_values[:5],labels=country_names[:5])
```

```python
plt.pie(country_values[:3],labels=country_names[:3])
```

```python
#calculate percentage - 2 places after decimal
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
#observation - zomato earns most in india, then USA, then UK
```

```python
final_df.columns
```

```python
# lets group by 3 features - which tell food rating
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
# food ratings are divided into zones - poor, avg, good, vgood, excellent - sorted on aggregate rating and color
# .size tells total of records
```

```python
# reset index and name it as a dataframe
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
# rename last column as Rating Count
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
# saving above as variable
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

# finding observations 
1. 2k records are non rated food
2. when rating is between 4.5 to 4.9 - excellant food
3. when rating 4.0 to 4.4 - very good
4. rating is 3.5 to 3.9 - good
5. rating bbetween 3.0 to 3.4- avg
6. when rating is 2.5 to 2.9 - avg
7. when rating is 2.0 to 2.4 - poor



```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
# for cleaned view , use matplot lib dimentions
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
# same conclusion derived from this barplot
# use hue to color according to you with your order of palette
sns.barplot(x="Aggregate rating", y="Rating Count", hue='Rating color', data=ratings, palette=['white','red','orange','yellow','green'])
```

```python
# since white is invisible , make it blue
sns.barplot(x="Aggregate rating", y="Rating Count", hue='Rating color', data=ratings, palette=['blue','red','orange','yellow','green'])
```

```python
# observations
1. non rated - blue - highest
2. maximum ratings are between 2.5 to 3.4 -orange
```

```python
## Count plot  - used for plotting categorical values
sns.countplot(x="Rating color",data=ratings, palette=['blue','red','orange','yellow','green'])
```
