Import both Datasets

```python
# read data 
df_menu=pd.read_csv('menus.csv', encoding='latin-1')        
# csv files have latin-1, read pandas documentation
df_menu.head(25)
```

```python
# read data 
df_foodPlaces=pd.read_csv('restaurants.csv', encoding='latin-1')  
df_foodPlaces.head(25
```

```python
#combine two tables on restaurant_id  
final_df=pd.merge(df_menu,df_foodPlaces, left_on='restaurant_id', right_on='id') 
final_df.head(30)
```

```python
final_df.isnull().sum()
```

```python
#Renaming some column of final dataset so we can distinguish category_x, category_y , name_x, name_y after merging
final_df=final_df.rename(columns={"category_x": "Food Type", "category_y": "Cuisines", "name_x": "Food Sub Type", "name_y": "Restaurant Name"})
final_df.head(25)
```

```python
## Remove NAN score and ratings
final_df.dropna(subset = ['score', 'ratings'])
```

```python
## Find out TOP 5 cuisines
cuisine_values=final_df.Cuisines.value_counts().values
cuisine_values
```

```python
# storing the sorted order by index
cuisines_names=final_df.Cuisines.value_counts().index
cuisines_names
```

```python
## plot a pie chart using above 2
plt.pie(cuisine_values[:5],labels=cuisines_names[:5])
```

```python
plt.pie(cuisine_values[:5],labels=cuisines_names[:5],autopct='%1.2f%%')
```

```python
final_df=final_df.rename(columns={"Food Type": "FoodType","Food Sub Type": "Category", "Restaurant Name": "RestaurantName"})
```

```python
final_df=final_df.dropna(subset = ['score', 'ratings','Category'])
final_df
```

```python
final_df.isnull().sum()
```

```python
## Find Top Food Types liked in US
final_df.groupby(['score','ratings']).size().reset_index()
```

```python
ratings=final_df.groupby(['score','ratings']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
# for cleaned view , use matplot lib dimentions
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,4)
sns.barplot(x="score", y="Rating Count", data=ratings)
```
